## Carregamento dos Dados

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('desafio_indicium_imdb.csv', index_col=0)

df.loc[966, 'Released_Year'] = 1995

df['Released_Year'] = df['Released_Year'].astype(np.float64)

df_certificate = df['Certificate'].str.get_dummies()
# df = pd.concat([df, df_certificate], axis=1)

df['Runtime'] = df['Runtime'].str.replace(' min', '').astype(np.float64)

df['No_of_Votes'] = df['No_of_Votes'].astype(np.float64)

df['Gross'] = df['Gross'].str.replace(',', '').astype(np.float64)

df_genres = df['Genre'].str.get_dummies(sep=', ')
# df = pd.concat([df, df_genres], axis=1)

df_director = df['Director'].str.get_dummies()
# df = pd.concat([df, df_director], axis=1)

df_actors = df[['Star1', 'Star2', 'Star3', 'Star4']].agg(', '.join, axis=1)
df_actors = df_actors.str.get_dummies(sep=', ')

df.head(5)
```

## Regressão de Overview

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report
import warnings

warnings.filterwarnings('ignore')
```

```python
df_classification = df.copy()

df_classification.dropna(subset=['Overview', 'Genre'], inplace=True)

df_classification['Genre'] = df_classification['Genre'].apply(lambda x: x.split(','))

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df_classification['Overview'])

mlb = MultiLabelBinarizer()
y = mlb.fit_transform(df_classification['Genre'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = OneVsRestClassifier(LogisticRegression())
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred, target_names=mlb.classes_))
```

## Regressão para Inferir Nota do IMDB

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import pickle
```

```python
def run_model(df_regression, model_name):
    x = df_regression.drop(columns=['IMDB_Rating'])
    y = df_regression['IMDB_Rating']

    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    y_pred = np.round(y_pred, 1)

    print("R²:", round(r2_score(y_test, y_pred), 3))
    print("MAE:", round(mean_absolute_error(y_test, y_pred), 3))
    print("MSE:", round(mean_squared_error(y_test, y_pred), 3))

    with open(f'{model_name}.pkl', 'wb') as arquivo:
        pickle.dump(model, arquivo)

    return y_test, y_pred
```

```python
df_regression = df.copy()
df_regression.fillna(0, inplace=True)

df_regression.drop(columns=['Series_Title', 'Certificate', 'Released_Year', 'Genre', 'Overview', 'Director', 'Star1', 'Star2', 'Star3', 'Star4'], inplace=True)

df_regression = pd.concat([df_regression, df_certificate], axis=1)
df_regression = pd.concat([df_regression, df_genres], axis=1)
df_regression = pd.concat([df_regression, df_director], axis=1)

print('Todos os Atributos:')
y_test, y_pred = run_model(df_regression, 'model_all_features')

print('\nSomente Released_Year, Runtime, Meta_score, No_of_Votes, Gross:')
df_regression = df.copy()
df_regression.fillna(0, inplace=True)
y_test, y_pred = run_model(df_regression[['Released_Year', 'Runtime', 'Meta_score', 'No_of_Votes', 'Gross', 'IMDB_Rating']], 'model_numeric_features')

print('\nSomente Meta_score:')
df_regression = df.copy()
df_regression.fillna(0, inplace=True)
y_test, y_pred = run_model(df_regression[['Meta_score', 'IMDB_Rating']], 'model_meta_score_feature')
```
